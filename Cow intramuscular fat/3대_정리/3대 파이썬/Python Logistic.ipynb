##### 근내지방도, 농장등급 변수변환
<p> 근내지방도 : {1 : (7~9), 0: (0~6)} </p>
농가등급 :  <br>
A: '1++C' '1++B' '1++A' <br>
B: '1+C' '1+B' '1+A' <br>
C: '1A' '1C' '2A' '1B' '2B' '3A' '2C' '3B' '3C' '등외'

```python
# 근내지방도 0, 1 변경
cow_3['target'] = cow_3['근내지방도'].apply(lambda x: 1 if x >= 7 else 0 if x < 7 else x)
print('변경 전 근내지방도 unique 및 counts')
print(cow_3.근내지방도.value_counts().sort_index())
print('\n')
print('변경 후 근내지방도 unique 및 counts')
print(cow_3.target.value_counts())
```

```python
# 농가등급 A, B, C로 변경
cow_3['farm_level'] = cow_3['농가등급'].apply(lambda x: 'A' if ((x == '1++A')|(x == '1++B')|(x == '1++C'))  
                                       else 'B' if ((x == '1+A')|(x == '1+B')|(x == '1+C')) else 'C')
```

```python
print('농가 등급 변경 전과 후 비교')
print('\n')
print(f'''변환 후 농가 등급 값: {cow_3[cow_3["farm_level"]=="A"]["farm_level"].unique()} 
변환 전 농가 등급 값: {cow_3[cow_3["farm_level"]=="A"]["농가등급"].unique()}''')
print('\n')
print(f'''변환 후 농가 등급 값: {cow_3[cow_3["farm_level"]=="B"]["farm_level"].unique()}
변환 전 농가 등급 값: {cow_3[cow_3["farm_level"]=="B"]["농가등급"].unique()}''')
print('\n')
print(f'''변환 후 농가 등급 값: {cow_3[cow_3["farm_level"]=="C"]["farm_level"].unique()}
변환 전 농가 등급 값: {cow_3[cow_3["farm_level"]=="C"]["농가등급"].unique()}''')
```

<h5> train_test_split, 7:3 </h5>

```python
X_3 = logit_df_slm.drop(['target'],axis=1)
Y_3 = logit_df_slm.target
X_train_3, X_test_3, Y_train_3, Y_test_3 = train_test_split(X_3, Y_3, test_size=0.3, random_state = 2023)
```

```python
# 데이터 다시 합쳐주기, index 똑같음
train_3 = pd.concat([X_train_3,Y_train_3], axis=1)
test_3 = pd.concat([X_test_3,Y_test_3], axis=1)
```

##### Modeling: Logistic Regression

```python
# Train set, Logistic Regression.fit()
model = sm.Logit.from_formula('''target ~ C(성별) + C(farm_level) + 도축개월 + 도체중 + 형매_도체중_평균 + 형매_등지방_평균 
    + 형매_근내지방_평균 + 형매_마릿수 + 어미형매_도체중_평균 + 어미형매_근내지방_평균 + 외할미형매_도체중_평균 
    + 외할미형매_등심단면적_평균 + 외할미형매_근내지방_평균''', train_3).fit()
print(model.summary())
print(f'모델 AIC:{model.aic}')
print('\n')
print(f'오즈비{np.exp(model.params)}')
print('\n')
```
