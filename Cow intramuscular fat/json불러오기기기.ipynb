<h4> json 형태로 읽어온 후 DataFrame으로 변경</h4> 
<p> 들어올 때 원래 Dict형태인 key:value로 [{'이표번호':002xx, '개월':29,---}로 들어오긴 하지만 값이 따로 분리되지 않고 key='이표번호', 나머지 값들이 다 value로 들어가 있음</p>
<p> DataFrame.from_dict 또는 DataFrame으로 변경해도 이표번호만 Column으로 취급되고 나머지 다 Row로 들어감</p>

```python
a = json.load(open('a0(44).txt','r', encoding = 'utf-8'))
print(type(a))
print('\n')
print(f'Key 변수:{a.keys()}')
print('\n')
print(f'Key에 따른 value:{a.values()}')
print('\n')
for i,j in a.items():
  print(f'key:{i}, value:{j}')
```

```python
# Open으로 불러온 json 
# 텍스트 파일 내용 그대로 들어옴
# print(a)
```

```python
b = pd.DataFrame.from_dict(a.copy())
print(b.head())
```

```python
b.to_csv('json파일_불러오기.csv', encoding='cp949')
```

<h4> Dict type DataFrame으로 변경 예시</h4> 
<p> Dict형태로 들어왔을 때 key와 value가 나눠져있음 </p>
<p> DataFrame으로 변경했을 경우 key가 변수로 들어가서 3*3 로 만들어짐 </p>

```python
a = {'num' : [1,2,3], 'name' :['a','b','c'], 'height' : [170,180,175], 'weight' : [70,80,60]}
print(a)
print('\n')
print(f'Key 변수:{a.keys()}')
print('\n')
print(f'Key에 따른 value:{a.values()}')
print('\n')
for i,j in a.items():
  print(f'key:{i}, value:{j}')
```

```python
b = pd.DataFrame(a)
ab = pd.DataFrame.from_dict(a)
print(b)
print('\n')
print(ab)
```
